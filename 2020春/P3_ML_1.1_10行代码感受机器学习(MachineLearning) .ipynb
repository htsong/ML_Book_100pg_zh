## 讨论

* 如果用更高次的多项式来拟合又会如何？

* 该如何比较多个拟合结果的优劣？


```python
#用一次多项式拟合，相当于线性拟合
z1 = np.polyfit(X, Y, 1)
z2 = np.polyfit(X, Y, 2)
p1 = np.poly1d(z1)
print (z1)  #[ 1.          1.49333333]
print (p1)  # 1 x + 1.493

print (z2)
print (np.poly1d(z2))
```

```python
plt.scatter(X, Y)
plt.plot(X, z1[0]*np.array(X) + z1[1])
plt.plot(X, z2[0]*np.power(np.array(X),2) + z2[1]*np.array(X) + z2[2])
plt.show()
```

```python
# 绘图案例 an example of matplotlib
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn
from IPython.display import display, clear_output
import time
x = np.linspace(0,5)
f, ax = plt.subplots()
ax.set_title("Bessel functions")

for n in range(1,10):
    time.sleep(1)
    ax.plot(x, jn(x,n))
    clear_output(wait=True)
    display(f)

# close the figure at the end, so we don't get a duplicate
# of the last plot
plt.close()
```
